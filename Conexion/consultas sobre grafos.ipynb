3. Algoritmo de Dijkstra
el algoritmo de Dijkstra es un m√©todo para encontrar el camino m√°s corto desde un nodo origen a todos los dem√°s nodos en un grafo ponderado (es decir, con pesos en las aristas), siempre que los pesos sean no negativos.

¬øpara qu√© sirve?
nos sirve para encontrar la ruta m√°s corta entre dos puntos (por ejemplo, en mapas, redes rutas de entrega).

y tambien nos sirve para calcular distancias m√≠nimas desde un nodo origen a todos los demas.

¬øcomo funciona?

1. Inicializa las distancias desde el nodo origen a todos los dem√°s como (infinito), excepto la del nodo origen, que es 0.

2. Marca todos los nodos como no visitados.

3. Elige el nodo no visitado con la menor distancia conocida.

4. Actualiza las distancias de sus vecinos si el nuevo camino es m√°s corto.

5. Marca el nodo como visitado

6. repite desde el paso 3 hasta que todos los nodos hayan sido vis√≠tados.


Algoritmo de Yen

El algoritmo de Yen es un m√©todo utilizado para encontrar los k caminos m√°s cortos (no necesariamente √∫nicos) entre un nodo origen y un nodo destino en un grafo dirigido o no dirigido, ponderado y sin ciclos negativos.


¬øPara qu√© sirve?

Para obtener varias rutas alternativas, no solo la m√°s corta.
Es √∫til en:

Redes de telecomunicaciones (rutas alternativas si una falla).

Sistemas de navegaci√≥n (opciones de rutas).

Log√≠stica y transporte.

¬øC√≥mo funciona?

Encuentra el camino m√°s corto desde el origen hasta el destino (usando Dijkstra, por ejemplo). Ese es el primer camino.

Luego, genera caminos "desviados" (llamados k-spur paths) tomando diferentes puntos intermedios del primer camino.

Para cada desviaci√≥n, se evita parte del camino anterior (para obtener una ruta diferente).

Se guarda cada nuevo camino en una lista ordenada.

Se repite el proceso hasta encontrar los k caminos m√°s cortos.

algunas ventajas del algor√≠tmo de yen pueden ser encontrar m√∫ltiples rutas eficientes.

Garantizar orden por costo (del m√°s corto al m√°s largo).

Es muy util para algor√≠tmos de ruteo resiliente.
