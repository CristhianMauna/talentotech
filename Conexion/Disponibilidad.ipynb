```python
# Confiabilidades individuales
A1 = 10000 /(10000 + 4)
A2 = 12000 /(12000 + 5)

#Paso 2 Nodo A en serie con ese subsistema

A3 = 0.999999999999999999999999999999999999999999999999999999999999999999 #Supon una confiabilidad distinta para A

# ----1. Caso en serie ----
# El sistema falla si uno falla
A_serie = A1 * A2 * A3
print(f"confiabilidad en serie (A_serie) = {A_serie}")

#---- 2. Caso en Paralelo ---
#EL sistema funciona si al menos uno funciona
A_paralelo = 1 - (1 - A1) * (1 - A2)
print(f"confiabilidad en paralelo (A_paralelo) = {A_paralelo}")

#-----3. Ejemplo mixto ----
#Nodo A en serie con subsistema paralelo entre B y C
# Paso 1: Calcular el paralelo B y C
sub_paralelo = 1 - (1 - A1) * (1 - A2)

A_mixto = A3 * sub_paralelo
print(f"confiabilidad en sistema mixto (A_mixto) = {A_mixto}")
```
