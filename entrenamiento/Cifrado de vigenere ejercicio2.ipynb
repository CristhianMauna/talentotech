```python
#!/usr/bin/env python
#-*- coding: utf-8  -*-

abc = 'abcdefghijklmnopqrstuvwxyz'

def cifrar(cadena, clave):
  texto_cifrado =''
  i = 0
  for letra in cadena:
    if letra in abc:
      suma = abc.find(letra) + abc.find(clave [i % len(clave)])
      modulo = suma % len(abc)
      texto_cifrado += abc[modulo]
      i += 1
    else:
      texto_cifrado += letra  # No cifra caracteres fuera del alfabeto
  return texto_cifrado

def descifrar(cadena, clave):
    texto_descifrado = ''
    i = 0
    for letra in cadena:
        if letra in abc:
            resta = abc.find(letra) - abc.find(clave[i % len(clave)])
            modulo = resta % len(abc)
            texto_descifrado += abc[modulo]
            i += 1
        else:
            texto_descifrado += letra  # No cifra caracteres fuera del alfabeto
    return texto_descifrado



def main():
    c = input('cadena a cifrar: ').lower()
    clave = input('clave: ').lower()
    texto_cifrado = cifrar(c,clave)
    print('texto cifrado:' , texto_cifrado)

    c = input('cadena a descifrar: ').lower()
    clave = input('clave:').lower()
    texto_descifrado = descifrar(c,clave)
    print('texto descifrado:' , texto_descifrado)




if __name__ == '__main__':
  main()
```
