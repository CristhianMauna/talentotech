```python
#funcion para verificar si un numero es primo usando lambda
es_primo =lambda n: n > 1 and all(map(lambda x: n % x != 0, range(2, int(n**0.5)+1)))

#Lista para almacenar los primeros 100 numeros primos
primos = []

#contador para probar números desde el 2 en adelante
numero = 2

#mientras no tenga los 100 primos, seguira buscando

while len(primos) < 100:
  if es_primo(numero):
    primos.append(numero)
  numero += 1

#imprime los 100 numeros primos
print("los 100 primeros números primos")
print(primos)
```
