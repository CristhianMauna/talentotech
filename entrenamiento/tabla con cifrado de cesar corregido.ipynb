```python
import pandas as pd
# descifrar con cifrado César
def descifrar_cesar(texto, clave):
    resultado = ""
    for char in texto:
        if char.isalpha():
            base = ord('A') if char.isupper() else ord('a')
            resultado += chr((ord(char) - base - clave) % 26 + base)
        else:
            resultado += char
    return resultado

# Función para detectar clave usando una palabra esperada
def detectar_clave(mensaje_cifrado, mensaje_real):
    for clave in range(26):
        posible = descifrar_cesar(mensaje_cifrado, clave)
        if posible.lower() == mensaje_real.lower():
            return clave
    return None  # No se encontró una clave válida

# Lista de mensajes cifrados y los mensajes que se espera recuperar
mensajes_cifrados = [
    "qwethsyhpqj cvbnmsdthp",
    "stykt",
    "ipmb ñihdo",
    "jkhdy nb nkbchf",
    "twp fqlas m mlrte",
    "mwlas qewdk",
    "fghi qnslty",
    "qmshywlevgyseb",
    "wmsths",
    "abpkuy",
    "psctfwmsb",
    "xmlqc",
    "wposlat",
    "bmwpka",
    "jpgkw",
    "wmslsifr",
    "pñlwmski",
    "wpsmt",
    "trmxstyqsi",
    "cmsitqplazw",

]

# Lista con los mensajes reales (descifrados correctamente)
mensajes_descifrados = [
    "bienvenidos nuevamente",
    "hello",
    "hola mundo",
    "balon de futbol",
    "hoy vamos a ganar",
    "julio cesar",
    "juan camilo",
    "ciberseguridad",
    "python",
    "laptop",
    "servidor",
    "cable",
    "trabajo",
    "clases",
    "virus",
    "firewall",
    "sotfware",
    "datos",
    "privacidad",
    "informacion",
]

#Palabras para usar como claves descritas
Palabras_clave = [
    'casa', 'hogar', 'sol', 'luna', 'estrella', 'cielo', 'mar', 'tierra', 'fuego', 'agua',
    'viento', 'nube', 'lluvia', 'trueno', 'montaña', 'valle', 'bosque', 'desierto', 'ciudad', 'pueblo'
]

# Detectar claves numericas
claves_detectadas = [
    detectar_clave(cifrado, claro)
    for cifrado, claro in zip(mensajes_cifrados,mensajes_descifrados)
]

#Reemplazar None con un indice alternativo (por si acaso)
claves_finales = [
    clave if clave is not None else i
    for i, clave in enumerate(claves_detectadas)
]

#Convertir las claves númericas en palabras
claves_palabras = [
    Palabras_clave[clave] if clave < len(Palabras_clave) else
    f"clave_{clave}"
    for clave in claves_finales
]

#Crear Dataframe
df = pd.DataFrame({
    "mensaje_cifrado": mensajes_cifrados,
    "clave_detectada": claves_palabras,
    "mensaje_descifrado": mensajes_descifrados,

})

#Imprimir Dataframe
print(df)


```
