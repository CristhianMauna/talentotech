```python
# Función que convierte texto a binario
def texto_a_binario(texto):
    binario = ''
    for caracter in texto:
        bin_caracter = format(ord(caracter), '08b')  # Convierte a binario de 8 bits
        binario += bin_caracter + ' '
    return binario.strip()

# Función que convierte binario a decimal
def binario_a_decimal(binario):
    binarios = binario.split()  # Divide la cadena en bloques de 8 bits
    decimales = [int(b, 2) for b in binarios]  # Convierte cada bloque a decimal
    return decimales

# Función opcional que convierte decimal a texto (para recuperar el mensaje original)
def decimal_a_texto(decimales):
    texto = ''.join([chr(d) for d in decimales])  # Convierte cada decimal a su carácter ASCII
    return texto

# Ejemplo de uso
if __name__ == "__main__":
    mensaje = input("Introduce un mensaje de texto: ")

    # Texto a binario
    binario = texto_a_binario(mensaje)
    print(f"\nTexto en binario:\n{binario}")

    # Binario a decimal
    decimales = binario_a_decimal(binario)
    print(f"\nTexto en formato decimal:\n{decimales}")

    # Opcional: Decimal a texto (recuperación del mensaje)
    mensaje_recuperado = decimal_a_texto(decimales)
    print(f"\nMensaje recuperado desde decimal:\n{mensaje_recuperado}")
```
