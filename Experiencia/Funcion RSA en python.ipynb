```python
import math

def rsa_k_method(p, q, m):
    # Paso 1: Calcular n y φ(n)
    n = p * q
    phi_n = (p - 1) * (q - 1)

    print(f"a. n = {p} * {q} = {n}")
    print(f"b. φ(n) = ({p}-1)*({q}-1) = {phi_n}")

    # Paso 2: Calcular k = φ(n) + 1
    k = phi_n + 1
    print(f"\n2. k = φ(n) + 1 = {k}")

    # Paso 3: Factorizar k para hallar posibles e, d
    print("\n3. Buscando factores de k = e*d con condiciones:")
    valid_pairs = []
    for e in range(2, phi_n):  # e debe estar entre 2 y φ(n)-1
        if k % e == 0:
            d = k // e
            if math.gcd(e, phi_n) == 1:
                valid_pairs.append((e, d))

    if not valid_pairs:
        print("   ❌ No se encontraron pares (e, d) válidos.")
        return

    # Elegimos el primer par válido
    e, d = valid_pairs[0]

    print(f"   ✅ Par válido encontrado: e = {e}, d = {d}")
    print(f"   Clave pública: (e, n) = ({e}, {n})")
    print(f"   Clave privada: (d, n) = ({d}, {n})")

    # Paso 4: Cifrar mensaje
    c = pow(m, e, n)
    print(f"\n4. Cifrado del mensaje m = {m}:")
    print(f"   c = m^e mod n = {m}^{e} mod {n} = {c}")

    # Paso 5: Descifrar mensaje
    m_decrypted = pow(c, d, n)
    print(f"\n5. Descifrado del mensaje cifrado c = {c}:")
    print(f"   m = c^d mod n = {c}^{d} mod {n} = {m_decrypted}")

    return {
        'Llave_publica': (e, n),
        'Llave_privada': (d, n),
        'Mensaje_cifrado': c,
        'Mensaje_descifrado': m_decrypted
    }

# Ejemplo de uso:
p = 11
q = 23
m = 3

rsa_k_method(p, q, m)

```
